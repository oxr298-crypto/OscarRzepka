```python
num1 = float(input(" Number 1: "))
num2 = float(input( "Number 2: "))
sign = input( "Choose a sign of operation (+, -, *, /): ")
#Inclass assignment 1/30/2026
if sign == "+":
  print(num1 + num2)
elif sign == "-":
  print(num1 - num2)
elif sign == "*":
  print(num1 * num2)
elif sign == "/":
  if num2 != 0:
    print(num1 / num2)
```
