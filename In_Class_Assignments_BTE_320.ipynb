```python
Sales = float(input("Enter the total sales: "))

COGS = .4 * Sales
GP = Sales - COGS

print(f"Sales: {Sales}$")
print(f"Cost of Goods Sold: {COGS}$")
print(f"Gross Profit: {GP}$")
```
